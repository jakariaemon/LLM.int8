```python
model_name = "vilt-b32-finetuned-vqa" #@param ["vilt-b32-finetuned-vqa"]
```

```python
import requests
import torch
from transformers import ViltForQuestionAnswering, ViltProcessor
from PIL import Image

model_id = f"dandelin/{model_name}"

# prepare image + question
url = "http://images.cocodataset.org/val2017/000000039769.jpg"
image = Image.open(requests.get(url, stream=True).raw)
text = "How many cats are there?"

processor = ViltProcessor.from_pretrained("dandelin/vilt-b32-finetuned-vqa")
model = ViltForQuestionAnswering.from_pretrained(model_id, device_map="auto", load_in_8bit=True)

# prepare inputs
encoding = processor(image, text, return_tensors="pt")
# cast the input in float16
encoding["pixel_values"] = encoding["pixel_values"].to(torch.float16)

# forward pass
outputs = model(**encoding)
logits = outputs.logits
idx = logits.argmax(-1).item()
print("Predicted answer:", model.config.id2label[idx])
```
